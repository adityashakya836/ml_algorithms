# EDA on water storage

```python
water_data = pd.read_csv('water_usage_data.csv')
water_data.head()
```

```python
water_data.describe()
```

```python
water_data.corr()
```

```python
unique_district = water_data['District'].unique()
unique_crop = water_data['Crop'].unique()
unique_irrigation_method =water_data['Irrigation Method'].unique()
```

```python
print(unique_district)
print(unique_crop)
print(unique_irrigation_method)
```

```python
for i in water_data.columns[3:]:
    plt.figure(figsize = (10,7))
    pH = water_data.groupby(['District']).sum()[i]
    plt.pie(pH,labels = unique_district,autopct = '%.2f%%')
    plt.title('Percentage of '+i+' in different soil types')
    plt.show()
```

```python
for i in water_data.columns[3:]:
    plt.figure(figsize = (10,7))
    pH = water_data.groupby(['Crop']).sum()[i]
    plt.pie(pH,labels = unique_crop,autopct = '%.2f%%',radius=1.5)
    plt.title('Percentage of '+i+' in different soil types')
    plt.show()
```

```python
for i in water_data.columns[3:]:
    plt.figure(figsize = (10,7))
    pH = water_data.groupby(['Irrigation Method']).sum()[i]
    plt.pie(pH,labels = unique_irrigation_method,autopct = '%.2f%%')
    plt.title('Percentage of '+i+' in different soil types')
    plt.show()
```

```python
arr3=[]
water_data[water_data["Water Consumption (liters/hectare)"]==water_data["Water Consumption (liters/hectare)"].min()]
for i in unique_district:
    x=water_data[water_data["District"]==i]
    sm=x["Water Availability (liters/hectare)"].sum()
    print(i,"=",sm)
    arr3.append(sm)
plt.figure(figsize=(25,7))
plt.title("District vs Water Availability ",fontsize=30)
plt.bar(unique_district,arr3,width=0.7,align="center",color="yellow")
plt.xlabel("District",color="b",fontweight="bold",fontsize=25)
plt.ylabel("Water Availability(liters/hectare)",color="b",fontweight="bold",fontsize=25)
```

```python
# Relation between water consumption and Water availability
plt.figure(figsize=(10,7))
plt.scatter(water_data["Water Consumption (liters/hectare)"],water_data["Water Availability (liters/hectare)"],c="indigo")
plt.title("Water consumption vs water availability",fontsize=25)
plt.xlabel("Water consumption",fontsize=15,c="r")
plt.ylabel("Water Avaulability",fontsize=15,c="r")
```
