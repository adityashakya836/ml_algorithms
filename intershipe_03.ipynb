####  DATA PREPROCESSING

```python
from sklearn.preprocessing import LabelEncoder
le_crop=LabelEncoder()
le_district = LabelEncoder()
le_season = LabelEncoder()
```

```python
crop_prod["District"]=le_district.fit_transform(crop_prod["District"])
```

```python
crop_prod["Crop"]=le_crop.fit_transform(crop_prod["Crop"])
```

```python
crop_prod["Season"]=le_season.fit_transform(crop_prod["Season"])
```

```python
crop_prod.head()
```

```python
x=crop_prod.iloc[:,:-1]
```

```python
x.head()
```

```python
y=crop_prod.iloc[:,-1]
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
x_train
```

#### APPLYING DECISION TREE ALGORITHM

```python
from sklearn.tree import DecisionTreeRegressor
lg=DecisionTreeRegressor(max_depth=13,min_impurity_decrease=0.01)
```

```python
lg.fit(x_train,y_train)
```

```python
y_pre=lg.predict(x_test)
```

```python
y_pre
```

```python
y_test
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,confusion_matrix
```

```python
print("r2_score= ",r2_score(y_pre,y_test))
print("  MAE = ",mean_absolute_error(y_pre,y_test))
```

#### APPLYING LINEAR REGRESSION ALGORITHM 

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
p=lr.predict(x_test)
```

```python
print(r2_score(p,y_test))
print(mean_absolute_error(p,y_test))
```
