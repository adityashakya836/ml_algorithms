# Machine Learning:

  * **Experience(E)** is the knowledge gained from data provided to the algorithm or model.
  * **Performance(P)** is the measure that indicates how welll a particular ML algorithm has performed the given task T using experience E.

# Types of Machine Learning Algorithms
* **Supervised ML algorithms** as the name indicates have a supervisor as teacher.
Apply Learning from past data to new data using labeled examples. Should be used when output of data in the training set is known.
For example, if we have input variable(s) X and output variable(Y), then the mapping function from input to output can given as, Y = f(X). This mapping function can be used to predict the output value for any new input after learning from the existing data.
Supervised learning  algorithms can be further classified into two categories

  1.   **Classification algorithms**: Classifies data into a particular group. Classification techniques predict discrete categories. The output will be based on what the model has learning in training phase. It can be used in medical imaging, speech recognition, credit scoring, prediction if an incoming email is authentic or spam. It can be used best if data can be tagged, categorised or sperated. 
      * Commonly used classification algorithms:
          * K-nearest neighbor(KNN)
          * Naive Bayes
          * Discriminant analysis

  2.   *Regression*: A regression algorithm predicts a real value. The output value is based on what the model has learned in its training phase.
        * In contrast to classification algorithms, regression predicts conitnuous values. Ex, The cost of a product, the value of a stock, changes in temperature or fluctuation in power demand.
        * Commonly used Regression Algorithms
          * Logisitic Regression, Neural Networks.
 
# Unsupervised Learning:
Trains the machine using information that is neither classified nor labeled. In this case, the ML algorithm works on that information without any guidance.

# Unsupervised ML algorithm 
Learns through observation and finding structures in the data. That is, the model automatically finds patterns and relationships in the dataset by creating clusters in it.

  1. ***Clustering*** : In clustering, the aim is to discover inherent groupings in the data or discover hidden patterns. It is one of the most useful unsupervised ML techniques as it finds similarity as well as relationship in the underlying data.
    * Commonly used Clustering algorithms include k-means, k-mediods, hierarhcical clustering, etc.
  2. ***Association Analysis*** : In association mining, the aim is to discover rules that describe large portions of data.
  3. ***Dimensionality Reduction*** : It is used to reduce the number of feature variables for the data set. It is done by selecting a set of prinicipal features. 
    * PCA, KNN and discriminant analaysis are some popular algorithms used for dimensionality reduction.
  4. ***Anomaly Detection*** : This technique is used to find out the occurrences of rare events or observations that generally do not occur. 

These algorithms use an iterative approach called deep learning to review data and arrive at conclusions.
These algorithms are used in image recognition, speech-to-text and natural language processing applications, predicting the probability of presence of a particular disease, etc.

# Data Rescaling

* Normalization : Refers to rescaling of floating-point values into the range 0 to 1. It is also known as Min-Max scaling. 
      X` = (X-Xmin)/Xmax-Xmin
where Xmax and Xmin are the maximum and the minimum values of the feature, respectively.
  * For very small values of X, the numerator will be 0 , and hence X` will be 0
  * For large values of X, the numerator is equal to the dnominator and thus the value of X` is 1
  * For X b/w minimum and the maximum value, the value X` is b/w 0 and 1

# Math behind Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
from  pandas.plotting import scatter_matrix
```

```python
url = "https://raw.githubusercontent.com/campusx-official/100-days-of-deep-learning/main/day3/placement.csv"
```

```python
data_pt = np.genfromtxt(url,delimiter = ",")
```

```python
data_pt
```

```python
input_val = []
output_val = []
for i in range(len(data_pt)):
  input_val.append(data_pt[i,0])
  output_val.append(data_pt[i,1])
```

```python
input_val
```

```python
output_val
```

```python
plt.scatter(input_val,output_val)
plt.xlabel("Input")
plt.ylabel("Output")
plt.show()
```

```python
#intial MSE
def mse_calc(slope, intercept, d_pt):
  tot_error = 0
  pred = []
  for i in range(len(d_pt)):
    x = d_pt[i,0]
    y = d_pt[i,1]
    pred.append(((slope*x)+intercept))
    tot_error += (y-((slope*x)+intercept))**2
  error = tot_error/float(len(d_pt))
  return error,pred
```

```python
# gradient descent
def grad_descent(s_slope, s_intercept, l_rate, iter_val, d_pt):
  for i in range(iter_val):
    int_slope = 0
    int_intercept = 0
    n_pt = float(len(d_pt))

    for i in range(len(d_pt)):
      x = d_pt[i,0]
      y = d_pt[i,1]
      int_intercept += -(2/n_pt)*(y-((s_slope*x)+ s_intercept ))
      int_slope += -(2/n_pt)*x*(y-((s_slope*x)+s_intercept))
    final_slope = s_slope-(l_rate*int_slope)
    final_intercept = s_intercept-(l_rate*int_intercept)
    s_slope = final_slope
    s_intercept = final_intercept
  return s_slope, s_intercept
```

```python
#defining slope and intercept value as 0
learning_rate = 0.0001
start_slope = 0
start_intercept = 0
iteration = 950

#intial run 
e_value, prediction = mse_calc(start_slope, start_intercept,data_pt)
print('initial error',e_value)

# graph
plt.scatter(input_val,output_val)
plt.plot(input_val,prediction, color = 'blue',linewidth = 3)
plt.xlabel("Input")
plt.ylabel("Output")
plt.show()
```

```python
# gradient descent
grad_slope, grad_intercept = grad_descent(start_slope, start_intercept,learning_rate,iteration,data_pt)
final_e_value, prediction = mse_calc(grad_slope, grad_intercept,data_pt)

```

```python
print('Slope',grad_slope)
print("Intercept",grad_intercept)
print("final_error",final_e_value)
```

```python
plt.scatter(input_val,output_val)
plt.plot(input_val,prediction, color = 'blue',linewidth = 3)
plt.xlabel("Input")
plt.ylabel("Output")
plt.show()
```
